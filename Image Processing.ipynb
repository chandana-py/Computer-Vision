# Computer Vision using OpenCV - Basic Image Processing Operations

### Objective
To perform basic image processing operations using OpenCV such as image reading,displaying,color conversion,grayscale conversion,resizing,scaling,flipping and cropping

##### 1.Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

##### 2. Read an Image

```python
img= cv2.imread("flowers.jpg")
```

##### 3. Display Image using OpenCV

```python
cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### 4. Displaying the image using matplotlib

```python
plt.imshow(img) # without  converting in rgb
plt.title("BGR")
plt.axis('off')
plt.show()
```

##### 5. Convert BGR to RGB

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) 
plt.title("RGB")
plt.axis('off')
plt.show()
```

##### 6. Convert BGR to GrayScale

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)) 
plt.title("BGR")
plt.axis('off')
plt.show()
```

##### 7. Convert Image to GrayScale

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("image",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(gray,cmap = "gray") 
plt.title("RGB Image")
plt.axis('off')
plt.show()
```

##### 8. Resizing  image

```python
resized_img = cv2.resize(img,(300,300))
cv2.imshow("image",resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
                         
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(resized_img,cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.show()
```

```python
# resizing with the scaling factor
# use fx and fy , when you want to scale an image relative to its current size
scaled_resized_img = cv2.resize(img,None,fx=0.2,fy=0.2)
cv2.imshow("image",scaled_resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt 
img= cv2.imread("flowers.jpg")
resized_img = cv2.resize(img,(300,300))
scaled_img = cv2.resize(img,None,fx=0.2,fy=0.2)
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
resized_rgb = cv2.cvtColor(resized_img,cv2.COLOR_BGR2RGB)
scaled_rgb = cv2.cvtColor(scaled_img,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(resized_rgb)
plt.title("Fixed Resize")

plt.subplot(1,3,3)
plt.imshow(scaled_rgb)
plt.title("Scaled Resize")
plt.show()
```

##### 9. Flipping image
 * Horizontal flip
 * Vertical flip
 * Both axes

```python
img = cv2.imread("flowers.jpg")
```

```python
horizontal = cv2.flip(img,1)
vertical = cv2.flip(img,0)
both = cv2.flip(img,-1)

plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(1,4,2)
plt.imshow(cv2.cvtColor(horizontal,cv2.COLOR_BGR2RGB))
plt.title("Horizontal")

plt.subplot(1,4,3)
plt.imshow(cv2.cvtColor(vertical,cv2.COLOR_BGR2RGB))
plt.title("Vertical")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(both,cv2.COLOR_BGR2RGB))
plt.title("both")
plt.show()
```

##### 10. Cropping the image

```python
import matplotlib.pyplot as plt 
import cv2
img= cv2.imread("flowers.jpg")
#check image shape
print("Image Shape:",img.shape)
h,w = img.shape[:2]

#define crop area safely inside image
row_start = int(h*0.2)
row_end = int(h*0.6)
col_start = int(w*0.2)
col_end = int(w*0.6)
cropped = img[row_start:row_end,col_start:col_end]

#display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(cropped,cv2.COLOR_BGR2RGB))
plt.title("cropped")
plt.show()
```

##### 11. Translation :means shifting the image
* scaling
* slanting
* shifting
* rotating

```python
import matplotlib.pyplot as plt 
img= cv2.imread("flowers.jpg")
rows,cols = img.shape[:2]
M = np.float32([[
    2,0,-20],[0,2,-30]])  #1:for scaling #0:for slanting # 50 : third column:for shifting
Translated = cv2.warpAffine(img,M,(cols,rows)) #warpAffine 
plt.imshow(cv2.cvtColor(Translated,cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.show()
```

##### 12. Rotating the image

```python
center = (cols//2,rows//2)
M = cv2.getRotationMatrix2D(center,10,1)
rotated = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.show()
```

##### 13. Drawing on image
* rectangle
* circle

```python
import cv2
import matplotlib.pyplot as plt


image=cv2.imread('flowers.jpg')
cv2.rectangle(image,(50, 50), (300, 200),(0, 255, 0),3)
cv2.circle(image,(50, 150),80,(255, 0, 0),  3)
cv2.putText(image,"My flower",(30, 300),cv2.FONT_HERSHEY_SIMPLEX,1,(0, 0, 255),2,cv2.LINE_AA )


plt.figure(figsize=(10,6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Rectangle + Circle + Text")
plt.show()
```

##### 14. Bitwise Operations on Image using OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

 
img1 = np.zeros((300, 300), dtype=np.uint8)
img2 = np.zeros((300, 300), dtype=np.uint8)



cv2.rectangle(img1, (50, 50), (200, 200), 255, -1)


cv2.circle(img2, (150, 150), 100, 255, -1)

bit_and = cv2.bitwise_and(img1, img2)
bit_or = cv2.bitwise_or(img1, img2)
bit_xor = cv2.bitwise_xor(img1, img2)
bit_not = cv2.bitwise_not(img2)

titles = ["Image 1", "Image 2", "AND", "OR", "XOR", "NOT"]

images = [img1, img2, bit_and, bit_or, bit_xor, bit_not]

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
```

* To create a simple image design using OpenCV drawing functions such as rectangles, circles, and text.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Create a white image
img = np.ones((500, 500, 3), dtype=np.uint8) * 255
# Draw a rectangle
cv2.rectangle(img, (50, 50), (200, 200), (255, 0, 0), 3)
# Draw a circle
cv2.circle(img, (350, 150), 60, (0, 255, 0), -1)
# Add your name
cv2.putText(img, "Chandana B", (120, 450),cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

## Task 2: Design a Simple Traffic Signal using OpenCV
#### Description
In this task, a blank image is created and three colored circles representing the red, yellow, and green traffic lights are drawn inside a rectangular signal box. A label is added below the drawing using OpenCV's text function, and the final output is displayed using Matplotlib.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = np.ones((500, 500, 3), dtype=np.uint8) * 255
cv2.rectangle(img, (180, 50), (320, 350), (0, 0, 0), 3) # Draw signal box
cv2.circle(img, (250, 100), 35, (0, 0, 255), -1) # Red light
cv2.circle(img, (250, 200), 35, (0, 255, 255), -1) # Yellow light
cv2.circle(img, (250, 300), 35, (0, 255, 0), -1)  # green light
cv2.line(img, (250, 350), (250, 470), (0, 0, 0), 5)  # Signal stand
cv2.putText(img, "Traffic Signal " , (130, 490),cv2.FONT_HERSHEY_SIMPLEX,0.8, (255, 0, 0), 2) # text
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
